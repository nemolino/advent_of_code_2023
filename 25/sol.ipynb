## PART 1

```python
import networkx as nx
```

```python
with open('input.txt') as f:
    input = list(map(lambda x: x.strip(), f.readlines()))

G = nx.Graph()

for l in input:
    src, targets = l.split(': ') 
    for t in targets.split():
        G.add_edge(src,t)

nx.draw(G)                    
```

```python
min_cut = nx.minimum_edge_cut(G, s=None, t=None, flow_func=None)
assert len(min_cut) == 3
```

```python
for edge in min_cut:
    G.remove_edge(edge[0],edge[1])

nx.draw(G)
```

```python
res = 1
for comp in nx.connected_components(G):
    res *= len(comp)
res
```
